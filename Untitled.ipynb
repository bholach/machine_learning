## Predicting Housing Prices

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

### 1.Loading data

```python
df = pd.read_csv('housing.csv')
```

```python
df.head()
```

```python
df.plot.scatter(x='RM',y='MEDV')
df.plot.scatter(x='PTRATIO',y='MEDV',color='green')
```

### 2.Spliting  Features and Labels 

```python
prices = df['MEDV']
prices.head()
```

```python
features = df.drop('MEDV', axis=1)
features.head()
```

### 3.Cleaning Data

```python
features.describe()
```

```python
len(features)
```

```python
features = features.dropna()
```

```python
len(features)
```

### 4.Splitting data into training and test data

```python
from sklearn.model_selection import train_test_split #importing train_test split module from sklearn
```

```python
X_train,X_test,y_train,y_test = train_test_split(features,prices,test_size=0.4,random_state=1)
```

```python
print("X train : ",len(X_train))
print("y train : ",len(y_train))
print("X test : ",len(X_test))
print("y test : ",len(y_test))
```

### 5.Train Model from data

```python
#importing linear regression estimator
from sklearn.linear_model import LinearRegression

model = LinearRegression(normalize=True) 

model.fit(X_train,y_train)  #fitting train data to model
```

```python
model.score(X_test,y_test) #gettign model score
```

### 6.Accuracy Score 

```python
from sklearn.metrics import accuracy_score
y_pred = model.predict(X_test)
#accuracy_score(y_test,y_pred)

print(y_test)
#print(y_pred)
```
